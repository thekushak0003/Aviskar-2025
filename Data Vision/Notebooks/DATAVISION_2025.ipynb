```python
corr_with_target = df[['Credited Units (1st Sem)',
                        'Approved Units (1st Sem)',
                        'Evaluated Units (1st Sem)',
                        'Enrolled Units (1st Sem)']].corrwith(df['Status_Code'])

print(corr_with_target)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_all = df.drop(columns=['Status_Code'])
y = df['Status_Code']

X_all = pd.get_dummies(X_all, drop_first=True)

X_reduced = df.drop(columns=[
    'Credited Units (1st Sem)',
    'Evaluated Units (1st Sem)',
    'Enrolled Units (1st Sem)',
     'Credited Units (2nd Sem)',
    'Evaluated Units (2nd Sem)',
    'Enrolled Units (2nd Sem)',
    'Status_Code'
])

X_reduced = pd.get_dummies(X_reduced, drop_first=True)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


Xtr, Xte, ytr, yte = train_test_split(X_all, y, test_size=0.3, random_state=42)
model1 = RandomForestClassifier(random_state=42, max_features='log2',max_depth=5).fit(Xtr, ytr)
acc1 = accuracy_score(yte, model1.predict(Xte))


Xtr2, Xte2, ytr2, yte2 = train_test_split(X_reduced, y, test_size=0.3, random_state=42)
model2 = RandomForestClassifier(random_state=42,max_features='log2',max_depth=5).fit(Xtr2, ytr2)
acc2 = accuracy_score(yte2, model2.predict(Xte2))

print("Accuracy with all features:", acc1)
print("Accuracy after dropping weak ones:", acc2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_all = df_new.drop(columns=['Status_Code'])
y = df_new['Status_Code']

X_all = pd.get_dummies(X_all, drop_first=True)

X_reduced = df_new.drop(columns=[
    'Credited Units (1st Sem)',
    'Evaluated Units (1st Sem)',
    'Enrolled Units (1st Sem)',
     'Credited Units (2nd Sem)',
    'Evaluated Units (2nd Sem)',
    'Enrolled Units (2nd Sem)',
    'Status_Code'
])

X_reduced = pd.get_dummies(X_reduced, drop_first=True)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


Xtr, Xte, ytr, yte = train_test_split(X_all, y, test_size=0.3, random_state=42)
model1 = RandomForestClassifier(random_state=42, max_features='log2',max_depth=4).fit(Xtr, ytr)
acc1 = accuracy_score(yte, model1.predict(Xte))


Xtr2, Xte2, ytr2, yte2 = train_test_split(X_reduced, y, test_size=0.3, random_state=42)
model2 = RandomForestClassifier(random_state=42,max_features='log2',max_depth=4).fit(Xtr2, ytr2)
acc2 = accuracy_score(yte2, model2.predict(Xte2))

print("Accuracy with all features:", acc1)
print("Accuracy after dropping weak ones:", acc2)
```
